```python
from torchvision import datasets
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
```

```python
rootPath ='/content/drive/MyDrive/SJSU_Fall_2024/CMPE_249_Intenlligent_Autonomous/HomeWork/HomeWork_1/CMPE249_HW_1'
#Donload Kitti Dataset
mykitti = datasets.Kitti(root=rootPath, train= True, transform = None, target_transform = None, download = False)
```

```python
# Function to draw bounding boxes on an image
def draw_boxes_on_image(image, label, output_path=None):

    draw = ImageDraw.Draw(image)

    # Load labels
    objects = label

    # Draw each bounding box
    for obj in objects:
        x1, y1, x2, y2 = obj['bbox']
        draw.rectangle([x1, y1, x2, y2], outline="red", width=2)
        draw.text((x1, y1 - 10), str(obj['type']), fill="red")  # Label above the box

    # Show or save the image
    if output_path:
        image.save(output_path)
        print(f"Image with boxes saved to: {output_path}")
    else:
        plt.figure(figsize=(10, 10))
        plt.imshow(image)
        plt.axis('off')
        plt.show()
```

```python
for i in range(5):
  img, targets=mykitti[i]
  draw_boxes_on_image(img,targets)
```
